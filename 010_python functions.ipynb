```python
# local variables
'''
-->> The variables  scope & life time will be with in the function defination
'''
def fun_name():
    m = 12  # this is local variable
    n = 34  # this is local variabe
    print "m local : ",m
    print "n local : ",n
fun_name()
print "m local : ",m
print "n local : ",n
```

```python
# global  variables
'''
-->> The variables  scope & life time will be through out ogf the program
'''
m = 12  # this is default global variable
n = 34  # this is default global  variabe
def fun_name():
    print "m global : ",m
    print "n global : ",n
    
fun_name()
print "m global : ",m
print "n global : ",n
```

```python
# global  variables
'''
-->> The variables  scope & life time will be through out ogf the program
'''
def fun_name():
    global m # creating the global variable
    global n
    m =123
    n = 345
    print "m local : ",m
    print "n local : ",n
    
fun_name()
print "m global : ",m
print "n global : ",n
```

```python
# recursive function
'''
A fun can call from itside of function defination
--->> A fun can calls from itself
'''
def fun1():
    print "fun1 started"
    print "fun1 stoped"
def fun2():
    print "fun2 started"
    fun1()
    print"fun2 stoped"
fun2()
```

```python
# recursive function
'''
A fun can call from itside of function defination
--->> A fun can calls from itself
'''
def fun1():
    print "fun1 started"
    fun1()
    print "fun1 stoped"
def fun2():
    print "fun2 started"
    fun1()
    print"fun2 stoped"
fun2()
```

```python
# recursive function
'''
A fun can call from itside of function defination
--->> A fun can calls from itself
'''
def fun1(a = 0):
    print "fun1 started"
    if(a <= 2):
        fun1(a+1)
    print "fun1 stoped"
def fun2():
    print "fun2 started"
    fun1()
    print"fun2 stoped"
fun2()
```
